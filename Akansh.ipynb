```python
#1 task
def max_of_three_numbers(a, b, c):
  return max(a, b, c)
print(max_of_three_numbers(8, 9, 10))
```

```python
#2 task
def sum(a, b, c, d, e):
  return(a + b + c + d + e)
print(sum(8, 2, 3, 0, 7))
```

```python
#3 task
def multiply(a, b, c, d, e):
  return(a * b * c * d * e)
print(multiply(8, 2, 3, -1, 7))
```

```python
#task 4
def reverse(string):
  string = text[::-1]
  print(string)
text = "lynnrose"
print(reverse(text))
```

```python
#task 5
def factorial(x):
  if x == 0:
    return 1
  else:
    return x * factorial(x-1)
x = int(input("give me the factorial: "))
print(factorial(x))
```
