```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
from google.colab import files
import mplcursors

#Upload the data file
upload = files.upload()
#Reading the data frame
data_frame = pd.read_csv('population_data.csv')
#Ask the user
a = ("Show the bar graph of this data to show the population trend")
b = ("Show the line graph of this data to show the population trend")
c = ("Extrapolate this data to show the population after 20 years")
choice = input("Choose between A, B, C: ")
#Plot according to choice
if choice == 'A':
  data_frame.plot(kind='bar', x='Year', y='Population')
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.title('Population Data')
  plt.show()
elif choice == 'B':
  data_frame.plot(kind='line', x='Year', y='Population')
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.title('Population Data')
  plt.show()
elif choice == 'C':
#Linear regression (X axis is now named year)
  x = data_frame[['Year']]
  y = data_frame['Population']
#Creating linear model and fitting into the data
  model = LinearRegression()
  model.fit(x, y)
#Extrapolate for 20 years in the future
#Reshaping the array from 1D to 2D with one column - reshape(-1, 1)
  years_ahead = np.array(range(2100, 2121)).reshape(-1, 1)
  population_predicted = model.predict(years_ahead)
#Plotting the regression line
  plt.plot(years_ahead, population_predicted, marker='o')
  plt.title('Linear Regression')
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.grid(True)
#Adding values of the points
#Enumerate allows you to iterate
#Annotating means it's adding those values
  for i, txt in enumerate(population_predicted):
    plt.annotate(txt, (years_ahead[i], population_predicted[i]))
#Using hover to see the coordinate values
  mplcursors.cursor(hover=True)
  plt.show()
else:
    print("ERROR!")
```
