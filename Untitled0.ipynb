```python
import random

class CricketGame:
    def __init__(self):
        self.runs_to_win = 0
        self.runs_scored = 0
        self.balls_remaining = 6
        self.wickets_remaining = 10
        self.overs_played = 0
        self.current_batsman = "Player"
        self.target_score = 0
        self.is_game_over = False

    def display_scoreboard(self):
        print("\nScoreboard:")
        print(f"{self.current_batsman}: {self.runs_scored}/{self.wickets_remaining}  ({self.overs_played} overs)")

    def play_ball(self):
        runs_on_this_ball = random.randint(0, 6)
        print(f"\n{self.current_batsman} is batting...")
        input("Press Enter to bowl...")

        if runs_on_this_ball == 0:
            print("Out! The batsman is dismissed.")
            self.wickets_remaining -= 1
        else:
            print(f"{runs_on_this_ball} runs scored on this ball.")
            self.runs_scored += runs_on_this_ball

        self.balls_remaining -= 1

        if self.balls_remaining == 0 or self.wickets_remaining == 0:
            self.end_inning()

    def end_inning(self):
        self.overs_played += 1
        self.balls_remaining = 6
        self.display_scoreboard()

        if self.wickets_remaining == 0 or self.runs_scored >= self.target_score:
            self.end_game()
        else:
            print("Next batsman is in.")
            self.current_batsman = f"Batsman {11 - self.wickets_remaining}"
            input("Press Enter to start next over...")

    def end_game(self):
        print("\nGame Over!")
        if self.runs_scored >= self.target_score:
            print(f"Congratulations! {self.current_batsman} wins the match.")
        else:
            print("Better luck next time. You couldn't achieve the target.")

        self.is_game_over = True

    def play_game(self):
        print("Welcome to the Cricket Game!")

        # Set up target score
        self.target_score = random.randint(50, 150)
        self.runs_to_win = self.target_score + 1

        print(f"\nYou need {self.runs_to_win} runs to win the match.")

        while not self.is_game_over:
            self.display_scoreboard()
            self.play_ball()

if __name__ == "__main__":
    game = CricketGame()
    game.play_game()



```
