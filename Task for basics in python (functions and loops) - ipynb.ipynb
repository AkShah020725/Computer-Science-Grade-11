```python
# Q1 - Python Exercise: Multiply all the numbers in a list
list = [1, 2, 3, 4, 5]
result = 1

for i in list:
    result = result * i

print(result)
```

```python
# Q3 - Write a Python function that takes a list and returns a new list with distinct elements from the first list. # Sample List : [1,2,3,3,3,3,4,5] # Unique List : [1, 2, 3, 4, 5]
# append is ading another value
sample_list = [1,2,3,3,3,3,4,5]
unique_list = []
for i in sample_list:
  if i not in unique_list:
    unique_list.append(i)
print(unique_list)
```
