```python
#Addition, Subtract, Division, Multiplication
import math as m
user_input = input("Which operation do you want: ")
x = float(input("Give me a number please: "))

if user_input == "add":
  y = float(input("Give me a number please: "))
  def add (x, y):
    return x+y
elif user_input == "subtract":
  y = float(input("Give me a number please: "))
  def subtract (x, y):
    return x-y
elif user_input == "multiply":
  y = float(input("Give me a number please: "))
  def multiply (x, y):
    return x*y
elif user_input == "division":
  y = float(input("Give me a number please: "))
  def division (x, y):
    return x/y
#Sin, cos, tan, power, log, - e^
if user_input == "sin":
    result = m.sin(x)
elif user_input == "cos":
    result = m.cos(x)
elif user_input == "tan":
    result = m.tan(x)
elif user_input == "power":
  y = float(input("Give me an exponent please: "))
  result = m.pow(x, y)
elif user_input == "log":
    result = m.log10(x)
elif user_input == "e":
    result = m.expm1(x)
print("The answer is", result)
```
