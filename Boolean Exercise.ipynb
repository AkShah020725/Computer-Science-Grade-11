```python
#1 Write a Python program to check if a given number is even or odd using boolean variables.
def even(number):
    return number % 2 == 0
x = int(input("Give me a number: "))
if even(x):
   print(f"The number is even")
else:
    print(f"The number is odd")
```

```python
#2 Write a program that calculates whether a person is eligible for voting based on their age using boolean conditions.
def required(age):
  return age >= 18
voter = int(input("What is your age? "))
if required(voter):
   print(f"You are able to vote")
else:
    print(f"You are not able to vote")
```
